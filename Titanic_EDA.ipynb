##Data Story

The Titanic dataset contains 891 passenger records and 10 columns.

Most columns are complete, but the Age column contains 177 missing values.

The dataset has both numerical and categorical variables.

The numerical columns include passenger information like Age, Fare, and Passenger Class, while categorical columns include Name, Sex, and Ticket.

This exploratory analysis provides a clear understanding of the dataset before applying machine learning models.
